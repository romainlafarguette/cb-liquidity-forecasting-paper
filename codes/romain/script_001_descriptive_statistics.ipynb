# Descriptive statistics

```python
import pandas as pd
import numpy as np
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="white", font_scale=1.5)
plt.rcParams['figure.figsize'] = (12*1.618, 12)
```

## Turkey

```python
japan_path = Path.cwd().parents[1] / 'data/raw/japan_cb_balance_sheet_daily.xlsx'
df_turkey = pd.read_excel(turkey_path, index_col=0, parse_dates=True)

short_names_d = {
    'Turkey, Balance Sheet & Flows of MFI Sector, Central Bank, Liabilities, Central Bank Reserve Money, Currency Issued, TRY':'currency_issued',
    'Turkey, Balance Sheet & Flows of MFI Sector, Central Bank, Liabilities, Other Central Bank Money, Try Deposits of Public Sector, TRY': 'government_deposits',
    'Turkey, Balance Sheet & Flows of MFI Sector, Central Bank, Assets, Foreign, Total, TRY': 'foreign_currency_assets',
    }

df_turkey = df_turkey.rename(columns=short_names_d)
df_turkey.index.name = 'date'
df_turkey = df_turkey.sort_index()
```

```python
_d = df_turkey.copy()
_d = _d/1e9

# Layout
fig, axs = plt.subplots(1, 3)
ax1, ax2, ax3 = axs.ravel()

fig.suptitle('Key Items of the Central Bank of Turkey Balance Sheet')

ax1.set_title('Currency issued')
ax1.plot(_d['currency_issued'], label='Currency issued')
ax1.legend(frameon=False, loc='best')
ax1.set_ylabel('Billions of Turkish Liras', labelpad=20)

ax2.set_title('Government deposits')
ax2.plot(_d['government_deposits'], label='Government deposits')
ax2.legend(frameon=False, loc='best')

ax3.set_title('Foreign currency assets')
ax3.plot(_d['foreign_currency_assets'], label='Foreign currency assets')
ax3.legend(frameon=False, loc='best')

fig.autofmt_xdate()

plt.show()
```

## Japan

```python
# Data loading
japan_path = Path.cwd().parents[1] / 'data/raw/japan_cb_balance_sheet_daily.xlsx'
df_japan = pd.read_excel(japan_path, index_col=0, parse_dates=True)

japan_short_names_d = {
    'Japan, Balance Sheet & Flows of MFI Sector, Central Bank, Liabilities & Net Assets, Banknotes, JPY':'currency_in_circulation',
    'Japan, Balance Sheet & Flows of MFI Sector, Central Bank, Liabilities & Net Assets, Deposits of the Government, JPY': 'government_deposits',
    'Japan, Balance Sheet & Flows of MFI Sector, Central Bank, Assets, Foreign Currency Assets, JPY': 'foreign_currency_assets',
    }

df_japan = df_japan.rename(columns=japan_short_names_d)
df_japan.index.name = 'date'
df_japan = df_japan.sort_index()
```

```python
_d = df_japan.copy()
_d = _d/1e12

# Layout
fig, axs = plt.subplots(1, 3)
ax1, ax2, ax3 = axs.ravel()

fig.suptitle('Key Items of the Central Bank of Japan Balance Sheet')

ax1.set_title('Currency in circulation')
ax1.plot(_d['currency_in_circulation'], label='Currency issued')
ax1.legend(frameon=False, loc='best')
ax1.set_ylabel('Trillions of Japanese Yen', labelpad=20)

ax2.set_title('Government deposits')
ax2.plot(_d['government_deposits'], label='Government deposits')
ax2.legend(frameon=False, loc='best')

ax3.set_title('Foreign currency assets')
ax3.plot(_d['foreign_currency_assets'], label='Foreign currency assets')
ax3.legend(frameon=False, loc='upper left')

fig.autofmt_xdate()

plt.show()
```
